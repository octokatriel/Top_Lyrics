```python

# Load text from local file
from google.colab import files
uploaded = files.upload()
```

```python
text = uploaded['Lyrics_top_200.txt'].decode("utf-8")
```

```python
#Split text by whitespace
import re
split_text = re.split('\s',text)
split_text = split_text
print(split_text)
```

```python
# Remove empty strings from list
while ('' in split_text):
  split_text.remove('')

print(split_text)
```

```python
# Remove unwanted punctuation

def remove_punctuation(string):
    punctuation = '''[]!"():;,.¿?'''
    for ele in string:
      for char in ele:
        if char in punctuation:
            string = string.replace(ele,"")
    return string

punc_text = [remove_punctuation(i) for i in split_text]
print(punc_text)
```

```python
# Lower case all words except for ones beginning with I
case_text = list(punc_text)
for i in range(len(punc_text)):
  if not punc_text[i].startswith("I"):
    case_text[i] =punc_text[i].lower()

print(case_text)

```

```python
# Count frequencies of all words
from collections import Counter
count = Counter(case_text)
print(count)
```

```python
# Save as csv
import csv
with open('All_Lyrics_Count.csv', encoding='utf-8-sig', mode='w') as fp:
    fp.write('Word,Count\n')  
    for tag, count in count.items():  
        fp.write('{},{}\n'.format(tag, count)) 
```

```python
# Count frequencies of words with 4 or more characters
fourchar = []
for i in range(len(case_text)):
  if len(case_text[i]) >= 4:
    fourchar.append(case_text[i])

fourcharcount = Counter(fourchar)
print(fourcharcount)
```

```python
with open('4_char_Lyrics_Count.csv', encoding='utf-8-sig', mode='w') as fp:
    fp.write('Word,Count\n')  
    for tag, count in fourcharcount.items():  
        fp.write('{},{}\n'.format(tag, count)) 
```
